## ðŸ“ˆ US Population and Life Expectancy

The following examples use data from the [Gapminder Five-Year Dataset](https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv), which includes country-level data collected every 5 years from 1952 to 2007. 


The dataset contains the following columns: `country`, `year`, `pop` (population), `lifeExp` (life expectancy), and `gdpPercap` (GDP per capita).



We filter it to the **United States only** using the `country` column, and create three simple plots with Matplotlib:

- Population over time (in millions)  
- Life expectancy over time (in years)  
- Both population and life expectancy together on one graph using two y-axes  

This gives a quick visual of how the US population and average lifespan have changed over time.


```python
#import gapminder dataset and matplotlib 
import pandas as pd 
import matplotlib.pyplot as plt 

#create dataframe from csv file (covered in dataframe notebook)
df = pd.read_csv('../data/gapminder-five-year-data.csv')

# Filter to a single country, United States
df_usa = df[df['country'] == 'United States']

year = df_usa['year']
# Convert population to millions
pop_millions = df_usa['pop'] / 1e6

# Plot
plt.plot(year, pop_millions)

#Define x, y, and labels. 
plt.xticks(range(1950, 2011, 10))
plt.yticks(range(125, 350, 25))
plt.xlabel('Year')
plt.ylabel('Population (millons)')
plt.title('Population of United States Over Time')

plt.show()
```

```python
#plot life expectancy in US over time 
df = pd.read_csv('../data/gapminder-five-year-data.csv')

# Filter to a single country, United States
df_usa = df[df['country'] == 'United States']

year = df_usa['year']
# Life expectancy 
life_exp = df_usa['lifeExp'] 

# Plot
plt.plot(year, life_exp)

#Define x, y, and labels. 
plt.xticks(range(1950, 2008, 10))
plt.xlabel('Year')
plt.ylabel('Life Expectancy (Years)')
plt.title('Avg Life Expectancy in the United States 1950-2007')
plt.grid(True)

plt.show()
```

```python
#plot population and life expectancy 

import matplotlib.pyplot as plt

#Add life expectancy to dataframe 
life_exp = df_usa['lifeExp']

fig, ax1 = plt.subplots()

# Plot population on left y-axis
ax1.plot(year, pop_millions, color='blue', label='Population (millions)')
ax1.set_xlabel('Year')
ax1.set_ylabel('Population (millions)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Create a second y-axis for life expectancy
ax2 = ax1.twinx()
ax2.plot(year, life_exp, color='green', label='Life Expectancy (years)')
ax2.set_ylabel('Life Expectancy (years)', color='green')
ax2.tick_params(axis='y', labelcolor='green')

# Optional: Add a title
plt.title('US Population and Life Expectancy Over Time')

# Show grid only on first y-axis
ax1.grid(True)

fig.tight_layout()
plt.show()
```
